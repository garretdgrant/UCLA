```python
from collections import deque
import csv

def bfs(start, end, x_capacity, y_capacity):
    path = []
    front = deque()
    front.append(start)
    visited = []
    visited.append(start)
    while(front):
#         Popping from the front of the queue
        current = front.popleft()
        x = current[0]
        y = current[1]
        path.append(current)
#         if current in visited:
#             continue
        if x == end:
            print("Found in " + str(len(path)) + " iterations!")
            return path

        # rule 1
        # (x, y) -> (x,3)
        if current[1] < y_capacity and ([current[0], y_capacity] not in visited):
            front.append([current[0], y_capacity])
            visited.append([current[0], y_capacity])

        # rule 3
        if current[0] > x_capacity and ([0, current[1]] not in visited):
            front.append([0, current[1]])
            visited.append([0, current[1]])

        # rule 4
        if current[1] > y_capacity and ([x_capacity, 0] not in visited):
            front.append([x_capacity, 0])
            visited.append([x_capacity, 0])

        # rule 5
        #(x, y) -> (min(x + y, x_capacity), max(0, x + y - x_capacity)) if y > 0
        if current[1] > 0 and ([min(x + y, x_capacity), max(0, x + y - x_capacity)] not in visited):
            front.append([min(x + y, x_capacity), max(0, x + y - x_capacity)])
            visited.append([min(x + y, x_capacity), max(0, x + y - x_capacity)])

        # rule 6
        # (x, y) -> (max(0, x + y - y_capacity), min(x + y, y_capacity)) if x > 0
        if current[0] > 0  and ([max(0, x + y - y_capacity), min(x + y, y_capacity)] not in visited):
            front.append([max(0, x + y - y_capacity), min(x + y, y_capacity)])
            visited.append([max(0, x + y - y_capacity), min(x + y, y_capacity)])
            
        #rule 7 - we can empty either pitcher
        #(x,y) -> (0, y) if x > 0
        if current[0] > 0 and [0,y] not in visited:
            front.append([0,y])
            visited.append([0,y])
        if current[1] > 0 and [x,y] not in visited:
            front.append([x,0])
            visited.append([x,0])

    return "Not found"



# start state: x = 0 , y = 0
start = [0, 0] 
end = 4
x_capacity = 5
y_capacity = 3


print(bfs(start, end, x_capacity, y_capacity))
```

```python
def min_dist_to_targ_and_init(frontier):
    min_cost_dist = math.inf
    min_targ_dist = math.inf
    golden_index = math.inf
    for i in range(len(frontier)):
        point = frontier[i]
        dist_to_targ = dist(point, (10,10))
        cost = dist_to_targ + dist(point, (0,0))

        if  min_cost_dist > cost:
            min_targ_dist = dist_to_targ
            min_cost_dist = cost
            golden_index = i
    
    return golden_index

def a_star():
    all_points = list(np.genfromtxt('HW1data.csv', delimiter=',', names=True))
    for i in range(len(all_points)):
        all_points[i] = tuple(all_points[i])
    path = []
    visited = deque()
    frontier = []
    visited.append((0,0))
    frontier.append((0,0))
    while frontier:
        min_dist_index = min_dist_to_targ_and_init(frontier)
        current_point = frontier.pop(min_dist_index)
        path.append(current_point)
        target_dist = dist(current_point, (10,10))
        if target_dist == 0:
            break
        
        for point in all_points:
            distance = dist(current_point, point)
           
            if distance < 1.3 and point not in visited:
                frontier.append(point)
                visited.append(point)
    print('Found in', len(path), 'iterations!', 'Total distance traveled: ', total_distance_traveled(path))
    print(path)
    path_plot(path, 'A* Path')
    return path

a_star()
```

```python
def total_distance_traveled(path):
    total_distance = 0
    for i in range(1, len(path)):
        total_distance += dist(path[i-1], path[i])
    return total_distance
```
