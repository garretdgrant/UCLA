## Problem 1

```python
import pandas as pd
import sklearn
```

### Importing data and splitting inputs/outputs

```python
heating_data = pd.read_csv('heating_load.csv')
X = heating_data[heating_data.columns[0:3]]
y = heating_data['heating load (BTU)']
```

```python
# Create test and validation data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

### Linear Regression using sklearn module

```python
from sklearn import linear_model
model_ols =  linear_model.LinearRegression()
model_ols.fit(X_train,y_train) 
coef = model_ols.coef_
intercept = model_ols.intercept_

print('Coef: ', coef, 'Intercept: ', intercept)
```

```python
from matplotlib import pyplot as plt
import math

y_prediction = model_ols.predict(X_test)
pred_vs_actual = pd.DataFrame(data={'Predictions': y_prediction, 'Actual': y_test})

mse = sklearn.metrics.mean_squared_error(y_test, y_prediction)

rmse = math.sqrt(mse)

plt.scatter(y_test, y_prediction)
plt.xlabel('Actual heating load')
plt.ylabel('Heating load predicted by model')
plt.rcParams["figure.figsize"] = (10,6) # Custom figure size in inches
plt.title("Heating Load from Data vs Heating Load from model")
print('Linear Regression rmse: ', rmse)
```

## Problem 1 part 2 - Lasso regularization

```python
# evaluate an lasso regression model on the dataset
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Lasso, LassoCV

# define model
# Lambda can be adjusted using the alpha variable
model = Lasso(alpha=1.0)
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force scores to be positive
scores = abs(scores)
print('Mean MAE: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
model.fit(X_train, y_train)
coef = model.coef_
intercept = model.intercept_
print('Coef: ', coef, 'Intercept: ', intercept)
```

```python
y_prediction = model.predict(X_test)

mse = sklearn.metrics.mean_squared_error(y_test, y_prediction)

rmse = math.sqrt(mse)

plt.scatter(y_test, y_prediction)
plt.xlabel('Actual heating load')
plt.ylabel('Heating load predicted by model')
plt.rcParams["figure.figsize"] = (10,6) # Custom figure size in inches
plt.title("Heating Load from Data vs Heating Load from model")

print('Lasso rmse: ', rmse)
```

### Tuning Lasso Hyperparameters

```python
from sklearn.model_selection import GridSearchCV

# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)

model = LassoCV(alphas=np.arange(0.01, 1, 0.01), cv=cv, n_jobs=-1)
# fit model
model.fit(X, y)
# summarize chosen configuration
print('alpha: %f' % model.alpha_)
```

```python
# define model
# Lambda can be adjusted using the alpha variable
model = Lasso(alpha=0.01)
# define model evaluation method
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force scores to be positive
scores = abs(scores)
print('Mean MAE: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))

model.fit(X_train, y_train)
coef = model.coef_
intercept = model.intercept_
print('Coef: ', coef, 'Intercept: ', intercept)
```

```python
y_prediction = model.predict(X_test)

mse = sklearn.metrics.mean_squared_error(y_test, y_prediction)

rmse = math.sqrt(mse)

plt.scatter(y_test, y_prediction)
plt.xlabel('Actual heating load')
plt.ylabel('Heating load predicted by model')
plt.rcParams["figure.figsize"] = (10,6) # Custom figure size in inches
plt.title("Heating Load from Data vs Heating Load from model")

print('Lasso rmse: ', rmse)
```

### preparing data

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)
scaler = StandardScaler()
X_std_train = scaler.fit_transform(X_train)
X_std_val = scaler.transform(X_val)
```

```python
param_grid = {
    "hidden_layer_sizes": [(6, 6, 6), (16, 8), (32,)],
    "activation": ["relu", "tanh"],
    "batch_size": [32, 64], 
    "learning_rate": ["constant", "adaptive"],
}
```

```python
ann_reg = MLPRegressor(max_iter=1000)
grid_search = GridSearchCV(ann_reg, param_grid, cv=5, 
                           scoring="neg_mean_squared_error",
                           return_train_score=True)
grid_search.fit(X_std_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
  print(np.sqrt(-mean_score), params)
```
